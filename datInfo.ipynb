```python
import numpy as np
import pandas as pd

#initial analysis on the data

df = pd.read_csv("data.csv")
del df['Screenshot']
print("The data contains " + str(df.shape[0]) + " rows and " + str(df.shape[1]) + " columns.\n")
print("The columns available are: " + str(df.columns.tolist()))
```

```python
#Get the domain names available in 'Style' column from the data which can be used to fill missing values

#Removed the values which was irrelevant
df.loc[df['Style']=='34', 'Style'] = ""
df.loc[df['Style']=='30', 'Style'] = ""

style = df['Style'].unique().tolist()
style.remove('')
style = [x for x in style if str(x) != 'nan']

print("There are " + str(df['Style'].isnull().sum()) + " rows which doesn't have values for their Style column. Have to be filled up!!\n")

print("The domain names available under Style column:")
print("(If needed, have to remove irrelevant values and add some if necessary)")
for i in range(len(style)):
    print(style[i])
```
